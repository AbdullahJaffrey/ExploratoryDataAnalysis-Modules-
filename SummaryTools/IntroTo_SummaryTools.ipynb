```python
from summarytools import dfSummary
```

```python
sns.get_dataset_names()
```

```python
pen = sns.load_dataset('penguins')
```

```python
dfSummary(pen)
```

```python
dfSummary(pen, is_collapsible = True)
```

```python
import pandas as pd

from summarytools import tabset

car_crashes = sns.load_dataset('car_crashes')

tabset({
    'penguin': dfSummary(pen).to_html(),
    'car_crashes': dfSummary(car_crashes).to_html()})
```
